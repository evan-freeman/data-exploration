# Let's Make Some Fake Data, and Visualize It

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from IPython.display import HTML
import seaborn as sns
```

```python
def f(x):
    return np.sin(x) * x

X = np.linspace(0, 100, 1000)
y = f(X)
```

```python
plt.figure(figsize = (10, 10))
sns.scatterplot(X, y)
plt.show()
```

```python
plt.figure(figsize = (10, 10))
plt.scatter(X, y)
plt.show()
```
