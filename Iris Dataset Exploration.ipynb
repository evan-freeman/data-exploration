Hello. My name is Evan Freeman. Here is a very famous dataset about Irises. Let's see what we can make of it using our skills as a data scientist.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from IPython.display import display

pd.options.display.max_columns = None
pd.options.display.max_rows = 30

iris = sns.load_dataset('iris')
```

<a name='exploring'></a>
## [Exploring the Data](#toc)

```python
display(iris)
```

### Let's calculate the ratio of length to width for sepal and petal.

```python
iris.insert(4, 'sepal_ratio', iris.sepal_length / iris.sepal_width)
iris.insert(5, 'petal_ratio', iris.petal_length / iris.petal_width)

display(iris)
```

<a name="plot"></a>
## [Plotting the Data](#toc)

```python
plt.figure(figsize = (10, 10))
sns.stripplot(x = 'species', y = 'petal_length', data = iris)
sns.set_context("notebook")
plt.show()
```

```python
plt.figure(figsize = (10, 10))
sns.swarmplot(x = 'species', y = 'petal_length', data = iris)
sns.set_context("notebook")
plt.show()
```

## Plotting Various Features

### Sepal Length vs Petal Length

```python
plt.figure(figsize = (10, 10))
sns.stripplot(x = 'sepal_length', y = 'petal_length', hue = 'species', data = iris)
sns.set_context("notebook")
plt.show()
```

### Sepal Width vs Petal Width

```python
plt.figure(figsize = (10, 10))
sns.stripplot(x = 'sepal_width', y = 'petal_width', hue = 'species', data = iris)
sns.set_context("notebook")
plt.show()
```
